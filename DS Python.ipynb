```python
# สร้าง function กำจัด outlier
def no_outliers(x):
     Q1 = x.quantile(0.25)
     Q3 = x.quantile(0.75)
     IQR = Q3 - Q1
     valid_values = ~((x < (Q1 - 1.5 * IQR)) |(x > (Q3 + 1.5 * IQR)))
     return valid_values
```

```python
# แสดงข้อมูล outlier
boolean_not_outlier =  no_outliers(df2["Sales"])

display(df2[~boolean_not_outlier])

# สร้าง df ใหม่
df2 = df2[boolean_not_outlier]
```

### K-Means clustering
จัดกลุ่มข้อมูลจำนวน k กลุ่ม หรือ cluster โดยใช้พื้นฐานทางสถิติ (Mean) ในการแบ่งกลุ่มข้อมูล เริ่มต้นจากสุ่มข้อมูล k จำนวนกระจายกัน เป็น k กลุ่ม แล้วเฉลี่ยข้อมูลจากทุกจุดที่อยู่ใกล้แต่ละกลุ่ม ได้เป็น จุดที่อยู่ตรงกลางข้อมูล หรือ centriod แล้วอาจจะทำซ้ำไปเรื่อยๆ จน centriod นิ่งไม่ขยับไปจุดไหนอีก จึงเป็น centriod ที่ดีที่สุด ในการแบ่งกลุ่มข้อมุลทั้ง k กลุ่ม

```python
# KMeans โดยใช้ sklearn
# ก่อนหน้าต้อง preprocessing ข้อมุลมาแล้ว แปลง scale และ เปลี่ยน target เป็นตัวเลข
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters = กำหนดจำนวนกลุ่ม, max_iter=จำนวนการทำซ้ำเพื่อหา centriod ,random_state=0)
df['cluster_label'] = kmean.fit_predict(df[[X1,X2,..]])

# plot clusters
sns.scatter(x="X1", y="X2", hue = 'cluster_labels', data=df3)
plt.label
plt.show()
```

```python
# การหา n_clusters ที่ดีทีสุดด้วย elbow method 
# ลองสัก 10 กลุ่ม และ หา distortion มาเปรียบเทียบดู error
# หากไม่รู้จำนวนกลุ่มที่แน่ชัด ควรทำก่อนเพื่อหา จำนวนกลุ่ม
distortion = []
for i in range(1,11):
    kmean = KMeans(n_clusters = กำหนดจำนวนกลุ่ม, max_iter=จำนวนการทำซ้ำเพื่อหา centriod ,random_state=0)
    kmean.fit(X)
    distortion.append(kmean.inertia_)
    
# plot elbow
plt.plot(range(1,11), distortion, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion(error)')
plt.show()
```

### Hierarchical clustering 
จัดกลุ่มข้อมุล โดยไม่กำหนดจำนวนกลุ่มจะแสดงผลเป็นจำนวนกลุ่มให้อัตโนมัติตามความเหมาะสม โดยใช้หลักการแบ่งข้อมูลเป็น hierarchical เป็นแผนผังแบ่งข้อมูลตามระยะจุดที่ใกล้เคียงกัน รวมกันเป็นเป็นชั้นๆ หรือ cluster ซึ่งแบ่งจำนวนข้อมูลจาก distence between data <เหมาะข้อมูลปริมาณไม่มาก และ ไม่รู้กลุ่มจำนวนข้อมูล ถ้ามากมักใช้ K-Mean> 

```python
from scipy.cluster.hierarchy import linkage, fcluster

# likeage คำนวนระยะห่างของข้อมูล เพื่อแบ่งข้อมูลเป็นกลุ่ม และ ใช้สร้าง dendrogram
distance_matrix = linkage(df[[X1,X2]]
                          , method='sigle/ward/complete'
                          ,metric='euclidean')

# fcluster กำหนด cluster label ให้ข้อมูล โดยใช้ระยะที่แบ่งจาก linkage 
df['cluster_label'] = fcluster(distance_matrix
                               , จำนวน clusterที่ต้องการ
                               , criterion = 'maxclust')

# plot cluster
sns.scatter(x="X1", y="X2", hue='cluster_labels',data=df)
plt.show()

# plot dendrogram 
# นำ distance_matrix จาก linkage มาทำแผนผัง dendrogram 
from scipy.cluster.hierarchy import dendrogram
dn = dendrogram(distance_matrix)
plt.show()
```
